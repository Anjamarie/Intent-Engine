```python
import pandas as pd
import numpy as np
shopper_df = pd.read_csv('online_shoppers_intention.csv')
```

```python
shopper_df.info()
```

```python
shopper_df['indecision_score'] = (shopper_df['ProductRelated_Duration'] / shopper_df['ProductRelated'].replace(0, 1)) * shopper_df['ExitRates']
```

```python
shopper_df.head()
```

```python
shopper_df['impulsivity_index'] = shopper_df['Administrative'] + shopper_df['Informational'] + shopper_df['ProductRelated']
shopper_df['impulsivity_index'] = shopper_df['impulsivity_index'] / (shopper_df['Administrative_Duration'] + shopper_df['Informational_Duration'] + shopper_df['ProductRelated_Duration']).replace(0, 1)

shopper_df.head()
```

```python
shopper_df['is_goal_oriented'] = ((shopper_df['Administrative'] == 0) & (shopper_df['Informational'] == 0)).astype(int)

print(shopper_df[['indecision_score', 'impulsivity_index', 'is_goal_oriented']].head())
```
